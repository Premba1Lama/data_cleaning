```python
#Import libraries

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Let's see if there is any missing data in our dataset and we will aslo look at the data types for each columns

df.info()
```

##### We can see from the above table that n_hos_beds have 8 datas missing

```python
#Missing value Treatment

df.n_hos_beds = df.n_hos_beds.fillna(df.n_hos_beds.mean())
```

```python
df.info()
```

##### Now we can see that n_hos_beds have no missing values 

```python
#Change data type of columns 

df['room_num'] = df['room_num'].astype('int64')
df['age'] = df['age'].astype('int64')
df['teachers'] = df['teachers'].astype('int64')
df['n_hos_beds'] = df['n_hos_beds'].astype('int64')
df['n_hot_rooms'] = df['n_hot_rooms'].astype('int64')
```

```python
df.head()
```

```python
#Let's check if there are any outliers 

df.describe()
```

##### 1. If you look at the mean and median value of independent variables, you will find out that for n_hot_rooms, the maximum value is significantly larger then 75 percentile or the mean value. So, data may be positively skewed or there may be some outliers in this data.

##### 2. For rainfall also, we can see that the minimum value is 3.000000, and the 25 percentile value is 28.000000 and then the maximum value is 60.000000. So, we can see that the data is either negatively skewed or there is an outlier in this data.

```python
#Outlier Treatment for n_hot_rooms

np.percentile(df.n_hot_rooms, [99])
```

```python
np.percentile(df.n_hot_rooms, [99]) [0]
```

```python
uv = np.percentile(df.n_hot_rooms, [99]) [0]
```

```python
df[df.n_hot_rooms > uv]
```

```python
df.n_hot_rooms[(df.n_hot_rooms > 3*uv)] = 3*uv
```

```python
#Outlier Treatment for rainfall

np.percentile(df.rainfall, [1])
```

```python
np.percentile(df.rainfall, [1]) [0]
```

```python
lv = np.percentile(df.rainfall, [1]) [0]
```

```python
df[df.rainfall < lv]
```

```python
df.rainfall[(df.rainfall < 0.3*lv)] = 0.3*lv
```

```python
#Let's check whether we have successfully treated the outliers 

df.describe()
```

##### Let's look at n_hot_rooms and rainfall. As we can see that for 'n_hot_rooms' maximum value is 45.000000 and the mean and median values are a lot closer than before. Similarly, for rainfall the lower value is now 6.000000 instead of 3.000000 and the median value is closer to the mean value.

```python
#Let's check which variables are most correlated to the dependent variable i.e price

df.corr()
```

##### 1. room_num has very high correlation coefficient with the price and we can say that this variable is going to be a very important variable for our analysis.
##### 2. Similarly, if you look at poor_prop this also has very high correlation coefficient with the price. So, we can say that price and percentage of poor population are highly negatively correlated and we can say that this variable is going to be a very important variable for our analysis. 
##### 3. You can see that n_hot_rooms has very low correlation with our dependent variable i.e price. So, this variable is not going to be important for our analysis.
##### 4. Rainfall also has a very low correlation coefficient with the price. So, this variable is not going to be important for our analysis.
##### 5. The correlation between air_qual and parks is 0.915544	. This means that these two independent variables are highly correlated. The high correlation between two independent variables leads to a problem known as multicollinearity. So, we have to delete one of these two variables. To  delete one of the variable we will check the correlation of these two variables with our dependent variable i.e price. For park we have correlation of  -0.391574 and for air_qual -0.429300. We delete the variable which has low correlation with our dependent variable. In this case, park has low correlation than air_qual. 



```python
#Let's delete park

del df['parks']
```
